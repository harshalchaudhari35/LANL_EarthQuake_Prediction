```python
data = {}
for filename in glob.glob('test/*.csv'):
    data[filename[:-4]] = pd.read_csv(filename)
```

```python
data['test/seg_00030f']['acoustic_data'].mean()
```

```python
all_keys=[]
for x in data.keys():
    all_keys.append(x)
## list of file names
```

```python
pair = {}
```

```python
for x in all_keys:
    print(x)
```

```python
for x in all_keys:
    normalize(data[x])
    a = check_quake(data[x])
    pair[x]=a
```

```python
def normalize(df):
    df['acoustic_data'] = df['acoustic_data'].swifter.apply(lambda x: abs(x)/df['acoustic_data'].mean())
    df.drop(df.loc[df['acoustic_data']==0].index, inplace=True)
    df['acoustic_data'] = df['acoustic_data'].swifter.apply(lambda x: math.log10(x))
    
def check_quake(df):
    return df['acoustic_data'].max()
#df.loc[df > 2.5]['acoustic_data'].max()

```

```python
with open('pair.json', 'r') as fp:
    pair = json.loads(fp.read())
```

```python
new_pair = {}
for key, value in pair.items():
    if value > 2.5:
        new_pair[key] = value
```

```python
# N = 10^a-bM 
# Ntot = 10^a
# therefore we have Ntot = 18
# Hence a = log(18) = 1.25
pair_with_zeros = {}
for key, value in pair.items():
    if value > 2.5:
        pair_with_zeros[key] = value
    else:
        pair_with_zeros[key] = 0
```

```python
pair_with_zeros
```

```python
## Count No. of Occurences
occurences = {}
for key, value in pair.items():
    occurences[key]= 10**(1.25-(1.0*value))
    
```

```python
occurences
```

```python
P_occurences = {}
for key, value in occurences.items():
    P_occurences[key]=value/18
```

```python
# Lambda
P_occurences
```

```python
# Inverse Lambda / Time_to_failure
inverse = {}
for key, value in P_occurences.items():
    inverse[key]=1/value
    
```
